```python
from math import comb

def dominant_probability(k, m, n):
    # Calculate the total number of organisms
    total_organisms = k + m + n
    
    # Calculate the total number of possible pairs
    total_pairs = comb(total_organisms, 2)
    
    # Calculate the probability for each type of mating pair
    prob_kk = (comb(k, 2) / total_pairs) * 1.0  # DD x DD
    prob_km = (k * m / total_pairs) * 1.0       # DD x Dd
    prob_kn = (k * n / total_pairs) * 1.0       # DD x dd
    prob_mm = (comb(m, 2) / total_pairs) * 0.75 # Dd x Dd
    prob_mn = (m * n / total_pairs) * 0.5       # Dd x dd
    
    # Calculate the overall probability
    overall_prob = prob_kk + prob_km + prob_kn + prob_mm + prob_mn
    
    return overall_prob

# Test the function
k = 15  # homozygous dominant
m = 16  # heterozygous
n = 27  # homozygous recessive

result = dominant_probability(k, m, n)
print(f"The probability that two randomly selected mating organisms will produce an individual possessing a dominant allele is {result:.5f}")
```
